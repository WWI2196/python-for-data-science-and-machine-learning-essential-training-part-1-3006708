## Data Filtering and Selection

```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
numbers_df = DataFrame(np.arange(0,126,3).reshape(14,3),
index = ['row 1', 'row 2', 'row 3','row 4','row 5','row 6','row 7','row 8','row 9','row 10','row 11','row 12','row 13','row 14'],
columns = ['column 1', 'column 2', 'column 3'])

numbers_df
```

```python
numbers_df.iloc[2,2]
```

```python
numbers_df.iloc[3,1] = 32
numbers_df
```

```python
numbers_df.iloc[[3,4,8],[1,2]]
```

#### Comparison operators (> < = <= => == !=) and Masking.

```python
check = numbers_df < 35
check
```

```python
check_1 = numbers_df >= 32
check_1
```

```python
numbers_df[check_1]
```

```python
numbers_df[numbers_df >= 32]
```

```python
numbers_df[numbers_df > 30] = 0
numbers_df
```

```python
numbers_df.iloc[2:6,1:]
```

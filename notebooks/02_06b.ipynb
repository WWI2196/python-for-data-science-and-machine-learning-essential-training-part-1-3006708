```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

### Concatenating data

```python
dataFrame_object = DataFrame(np.arange(49).reshape(7,7))
dataFrame_object
```

```python
dataFrame_objects_2 = DataFrame(np.arange(72).reshape(9,8))
dataFrame_objects_2
```

```python
pd.concat([dataFrame_object,dataFrame_objects_2],axis=1)
```

```python
pd.concat([dataFrame_object,dataFrame_objects_2],axis=0)
```

### Transforming data
#### Dropping data

```python
dataFrame_object.drop([0,2])
```

```python
dataFrame_object.drop([0,2],axis=1)
```

#### Adding data

```python
series_object = Series(np.arange(8))
series_object.name = 'added_variable'
series_object
```

```python
variable_added = DataFrame.join(dataFrame_objects_2,series_object)
variable_added
```

```python
added_datatable = pd.concat([variable_added,variable_added],ignore_index=False)
added_datatable
```

```python
added_datatable = pd.concat([variable_added,variable_added],ignore_index=True)
added_datatable
```

#### Sorting data

```python
data_sorted = dataFrame_object.sort_values(by=[5],ascending=[False]) # sorting data by column 5
data_sorted
```
